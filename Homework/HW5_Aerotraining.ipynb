```python
import pandas as pd
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

Since the data contains some NaN value, we will clean it by removing any row containing that value from our train data. But, first we will drop 'Cabin' since it is unnecessary and contains too many NaN value.


```python
clean_df = df.drop('Cabin',axis=1).dropna(axis=0,how="any")
```

```python
clean_df.info()
```

```python
pd.options.mode.chained_assignment = None

clean_df['Sex'].replace({
    "male":0, 
    "female":1},inplace=True)
clean_df
```

```python
#define the features used for the model
X = clean_df[['Sex','Age','Pclass','Fare']]

y = clean_df['Survived']
```

```python
#normalize the features used
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit(X).transform(X)

#split the given data to train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.3,random_state = 1)
```

```python
#build the logistic regression model based on train dataset
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(X_train,y_train)

#...and Support Vector Machine model for comparison
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)

#...also Decision Tree model for comparison
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
```

```python
#check the 'accuracy' of models when applied to its train dataset
print(logit.score(X_train,y_train))
print(svc.score(X_train,y_train))
print(tree.score(X_train,y_train))
```

```python
#check the accuracy of the models using test data
print(logit.score(X_test,y_test))
print(svc.score(X_test,y_test))
print(tree.score(X_test,y_test))
```

We will use the models to predict from a set of data of the other passenger 

```python
data_to_predict = pd.read_csv("test.csv")
data_to_predict.head()
```

```python
data_to_predict.info()
```

```python
#clean the data first so the model could predict whether the passenger survive or not

#since the model uses 4 features to make prediction, so we remove them if the data is not available
data_to_predict = data_to_predict.dropna(axis=0,how='any',subset=['Sex','Age','Pclass','Fare'])

data_to_predict['Sex'].replace({"male":0,"female":1},inplace=True)
data_to_predict.info()
```

```python
#normalize the data
X_predict = data_to_predict[['Sex','Age','Pclass','Fare']]
X_predict_norm = scaler.transform(X_predict)

#predicting..
y_predict_logit = logit.predict(X_predict_norm)
y_predict_svc = svc.predict(X_predict_norm)
y_predict_tree = tree.predict(X_predict_norm)
```

```python
#summary of predicted values from all models
predicted_data = data_to_predict.copy()
predicted_data['Survived_LogitR'] = y_predict_logit
predicted_data['Survived_SVC'] = y_predict_svc
predicted_data['Survived_Tree'] = y_predict_tree
predicted_data
```
